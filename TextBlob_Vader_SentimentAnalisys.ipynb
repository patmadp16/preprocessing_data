### Read Dataset

#PreProcessing Text

####Case Folding
<p align='justify'>Mengubah huruf besar menjadi huruf kecil</p>

```python
#------Case Folding------
komentar['CaseFolding']=komentar['Text'].str.lower()
komentar[['CaseFolding','Text']].head(5)
```

#### Hapus Tanda RT dan @ beserta nama mention

```python
import re
def remove_pattern(text,pattern):
  r = re.findall(pattern,text)
  for i in r:
    text = re.sub(i, '' , text)
  return text

komentar['CaseFolding']=np.vectorize(remove_pattern)(komentar['CaseFolding'],"#[\w]*")
komentar['CaseFolding']=np.vectorize(remove_pattern)(komentar['CaseFolding'],"@[\w]*")
komentar[['CaseFolding','Text']].head(5)
```

#### Menghapus URL dalam Tweet

```python
#remove url
import re
def clean_links(text):
  text = re.sub(r"http\S+", "", text)
  return text

komentar['CaseFolding']=komentar['CaseFolding'].apply(clean_links)
komentar[['CaseFolding','Text']].head(5)
```

#### Remove Char, Punctuation, Emoji



```python
import string 
import re 

from nltk.tokenize import word_tokenize 
from nltk.probability import FreqDist


string.punctuation

def remove_char(text):
  text = text.replace("b'","")
  text = text.replace("-"," ")
  return text.replace(":", " ")

#Menghapus tanda baca
def remove_punctuation(text):
    punctuation_comments="".join([i for i in text if i not in string.punctuation])
    text.replace(punctuation_comments," ")
    return punctuation_comments

#Meghapus emoji
def deEmojify(text):
    return text.encode('ascii', 'replace').decode('ascii')

#Menghapus angka
def remove_number(text):
    return  re.sub(r"\d+", "", text)

komentar['CaseFolding'] = komentar['CaseFolding'].apply(remove_char).apply(deEmojify)
komentar[['CaseFolding','Text']].head(10)
```

```python

#text cleaning
import re

def clean_text(text):
  #Remove \n, \t, \r
  text = re.sub(r"\\n","",text)
  text = re.sub(r"\\r"," ",text)
  text = re.sub(r"\\t"," ",text)

  text = re.sub("\s+"," ",text)

  #Mengganti &amp dengan &
  text = re.sub(r"&amp;"," ",text)
  
  text=text.lower()

  #Remove tanda baca
  text = re.sub(r"[?!.,;:#@_+()|\/<>]","",text)
  text = text.replace("https", '').replace("http",'')

  return text
  
def remove_emoticon(text):
  text = text.encode('ascii',"replace").decode('ascii')
  return text  

komentar['CaseFolding'] = komentar['CaseFolding'].apply(clean_text)

komentar['CaseFolding']= komentar['CaseFolding'].str.replace("[^a-zA-Z#][\a]"," ")

komentar[['CaseFolding','Text']].head(5) 
```

```python
# Menghapus double word
from collections import OrderedDict
komentar['CaseFolding'] = (komentar['CaseFolding'].str.split().apply(lambda x: OrderedDict.fromkeys(x).keys()).str.join(' '))

komentar[['CaseFolding','Text']].head()

```

### Tokenizing


```python
#defining function for tokenization
import re
import nltk
nltk.download('punkt')

def word_tokenize_wrapper(text):
    return word_tokenize(text)

#applying function to the column
komentar['tokenizing']= komentar['CaseFolding'].apply(word_tokenize_wrapper)
komentar[['tokenizing','CaseFolding']].head(5)
```

#### Normalisasi slang word
<p align='justify'>Merubah kata-kata alay atau singkatan menjadi kalimat yang baku</p>

```python
normalizad_word = pd.read_excel("/content/drive/MyDrive/NLP/Normalisasi_word.xlsx")

normalizad_word_dict = {}

for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

komentar['tokenizing'] = komentar['tokenizing'].apply(normalized_term)

komentar[['tokenizing','CaseFolding']].head(20)
```

### Label Using TextBolb

```python
komentar['transalted'] = komentar['transalted'].astype('str')

def get_polarity(text):
    return TextBlob(text).sentiment.polarity

komentar['Polarity'] = komentar['transalted'].apply(get_polarity)
komentar[['transalted','Polarity']].head()

komentar['Sentiment_Type']=''
komentar.loc[komentar.Polarity>0,'Sentiment_Type']='POSITIVE'
komentar.loc[komentar.Polarity==0,'Sentiment_Type']='NEUTRAL'
komentar.loc[komentar.Polarity<0,'Sentiment_Type']='NEGATIVE'

print(komentar['Sentiment_Type'].value_counts())
#twit_post = komentar[komentar["Sentiment_Type"] == "NEGATIVE"]

komentar[['transalted','Sentiment_Type','hasil']].head(5)
```

### Label Using Vader

```python
sid = SentimentIntensityAnalyzer()
komentar['scores'] = komentar['transalted'].apply(lambda transalted: sid.polarity_scores(transalted))
komentar['compound']  = komentar['scores'].apply(lambda score_dict: score_dict['compound'])

komentar['sentiment_type_Vanbes']=''
komentar.loc[komentar.compound>=0,'sentiment_type_Vanbes']='POSITIVE'
komentar.loc[komentar.compound==0,'sentiment_type_Vanbes']='NEUTRAL'
komentar.loc[komentar.compound<0,'sentiment_type_Vanbes']='NEGATIVE'

print(komentar['sentiment_type_Vanbes'].value_counts())

#twit_negatif = komentar[komentar["sentiment_type_Vanbes"] == "NEGATIVE"]


komentar[['transalted','sentiment_type_Vanbes','hasil']].head(5)
```
